```python
import math

def minimax(level, idx, is_max, leaf_values, max_level):
    """
    Minimax algorithm to evaluate the best possible value
    for the maximizing player.
    
    level       → current depth of the node
    idx         → index of the current node
    is_max      → True = max player's turn, False = min player's turn
    leaf_values → final score values at terminal nodes
    max_level   → depth at which recursion stops
    """

    if level == max_level:
        return leaf_values[idx]

    if is_max:
        left = minimax(level + 1, idx * 2, False, leaf_values, max_level)
        right = minimax(level + 1, idx * 2 + 1, False, leaf_values, max_level)
        return max(left, right)

    else:
        left = minimax(level + 1, idx * 2, True, leaf_values, max_level)
        right = minimax(level + 1, idx * 2 + 1, True, leaf_values, max_level)
        return min(left, right)

end_scores = [3, 5, 2, 9, 3, 5, 2, 9]   

final_depth = int(math.log(len(end_scores), 2))   
print("Best value chosen by Minimax =", end=" ")
print(minimax(0, 0, True, end_scores, final_depth))
```
